```python
import PyPDF2  
    
# creating a pdf file object  
pdfFileObj = open('C:/Users/Heng1222/Desktop/MPPE1113 Cognitive Psychology - Data 2019-2020-20210126T023625Z-001/MPPE1113 Cognitive Psychology - Data 2019-2020/SEM2-1920/REFLECTION INDIVIDUAL ASSIGMENT TAN SHIH MIN.pdf', 'rb')  
    
# creating a pdf reader object  
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)  
    
# printing number of pages in pdf file  
print(pdfReader.numPages)  

for i in range (0,pdfReader.numPages):
    # creating a page object  
    pageObj = pdfReader.getPage(i)  

    # extracting text from page  
    print(pageObj.extractText())  

    # closing the pdf file object  
pdfFileObj.close()  
```
