```python
import pandas as pd
# Read train data
df = pd.read_csv('C:/Users/Heng1222/sst_train.txt', sep='\t', header=None, names=['truth', 'text'],encoding='latin-1')
df['truth'] = df['truth'].str.replace('__label__', '')
df['truth'] = df['truth'].astype(int).astype('category')
```

```python
import requests
def translate(text):
    src='en'##english, can chg to "auto" for language detection
    dest='ms' #malay
    url = "https://clients5.google.com/translate_a/t?client=dict-chrome-ex&sl="+src+"&tl="+dest+"&q=" + text
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.104 Safari/537.36'
    }

    try:
        request_result = requests.get(url, headers=headers).json()

        #print(request_result)
        #print('-------')
        #print('[In Malay]: ' + request_result['alternative_translations'][0]['alternative'][0]['word_postproc'])
        result=request_result['alternative_translations'][0]['alternative'][0]['word_postproc']
        #print('[Language Dectected]: ' + request_result['src'])
    except:
         pass
    return result
```

```python
column_names = ["sentence","label"]
df_new = pd.DataFrame(columns = column_names)
for i in range(0,len(df['text'])):
    #translated=[]
    text=df['text'][i]
    print(text)
    translated=translate(text)
    print("Malay:",translated)
    df2 = pd.DataFrame({"sentence":[translated], "label":[df['truth'][i]]}) 
    df_new = df_new.append(df2, ignore_index = True)
```

```python
df_new.to_csv(r'C:/Users/Heng1222/Desktop/sst-5_malay.csv', index = False)
```
